```python
from collections import deque
import pandas as pd

def fifo_pnl(trade_history_df: pd.DataFrame, current_market_price: float):
    """
    Calculate FIFO-based realized and unrealized P&L from a trade history.

    Parameters
    ----------
    trade_history_df : pd.DataFrame
        DataFrame with at least ['Action', 'Price', 'Quantity'].
        Buys should have Quantity > 0; Sells should have Quantity < 0 (signed quantities).
        If a 'Date' column exists, it will be parsed and used to sort trades chronologically.
    current_market_price : float
        Current market price for marking open positions to market (unrealized P&L).

    Returns
    -------
    dict:
        'summary': key totals and metrics
        'match_ledger': per-lot matching details (sell  buy)
        'trade_summary': realized P&L per trade row
    """

    # Copy to avoid modifying original DataFrame
    trades_df = trade_history_df.copy()

    # Ensure numeric types
    trades_df["Price"] = pd.to_numeric(trades_df["Price"], errors="coerce")
    trades_df["Quantity"] = pd.to_numeric(trades_df["Quantity"], errors="coerce")

    # Sort chronologically if Date column exists
    if "Date" in trades_df.columns:
        trades_df["Date"] = pd.to_datetime(trades_df["Date"], errors="coerce")
        trades_df = trades_df.sort_values("Date").reset_index(drop=True)
    else:
        trades_df = trades_df.reset_index(drop=True)

    # FIFO lots: each entry = [shares_remaining, purchase_price, buy_trade_index, buy_trade_date]
    fifo_lots = deque()

    cumulative_realized_pl = 0.0  # sum of all realized profits/losses
    lot_match_records = []        # per-lot matches (each sell vs earlier buys)
    trade_summary_records = []    # per-trade realized P&L

    # Process each trade in order
    for trade_index, trade_row in trades_df.iterrows():
        trade_quantity = float(trade_row["Quantity"])
        trade_price = float(trade_row["Price"])
        trade_date = trade_row["Date"] if "Date" in trades_df.columns else None

        if trade_quantity > 0:
            # BUY trade: add to FIFO queue
            fifo_lots.append([trade_quantity, trade_price, trade_index, trade_date])
            trade_summary_records.append({
                "row": trade_index,
                "date": trade_date,
                "action": "Buy",
                "price": trade_price,
                "quantity": trade_quantity,
                "realized_for_trade": 0.0
            })

        elif trade_quantity < 0:
            # SELL trade: match shares against FIFO lots
            shares_to_sell = -trade_quantity  # positive number
            realized_this_trade = 0.0

            while shares_to_sell > 0 and fifo_lots:
                lot_quantity, lot_price, lot_index, lot_date = fifo_lots[0]

                # Number of shares taken from this lot
                shares_from_lot = min(lot_quantity, shares_to_sell)

                # Profit/Loss from these shares
                match_pl = (trade_price - lot_price) * shares_from_lot
                cumulative_realized_pl += match_pl
                realized_this_trade += match_pl

                # Record this match
                lot_match_records.append({
                    "sell_row": trade_index,
                    "sell_date": trade_date,
                    "sell_price": trade_price,
                    "buy_row": lot_index,
                    "buy_date": lot_date,
                    "buy_price": lot_price,
                    "quantity_matched": shares_from_lot,
                    "pl_on_match": match_pl
                })

                # Adjust remaining quantities
                lot_quantity -= shares_from_lot
                shares_to_sell -= shares_from_lot

                if lot_quantity == 0:
                    fifo_lots.popleft()  # lot fully consumed
                else:
                    fifo_lots[0][0] = lot_quantity  # update qty in lot

            trade_summary_records.append({
                "row": trade_index,
                "date": trade_date,
                "action": "Sell",
                "price": trade_price,
                "quantity": trade_quantity,
                "realized_for_trade": realized_this_trade
            })

        else:
            # No position change (Quantity = 0)
            trade_summary_records.append({
                "row": trade_index,
                "date": trade_date,
                "action": "Flat",
                "price": trade_price,
                "quantity": trade_quantity,
                "realized_for_trade": 0.0
            })

    # --- Position metrics ---
    open_quantity = sum(lot_qty for lot_qty, _, _, _ in fifo_lots)
    total_cost_basis = sum(lot_qty * lot_px for lot_qty, lot_px, _, _ in fifo_lots)
    average_cost_basis = (total_cost_basis / open_quantity) if open_quantity else 0.0
    market_value = open_quantity * current_market_price
    unrealized_pl = market_value - total_cost_basis
    total_pl = cumulative_realized_pl + unrealized_pl

    # DataFrames for detailed outputs
    match_df = pd.DataFrame(lot_match_records)
    trade_df = pd.DataFrame(trade_summary_records).sort_values("row").reset_index(drop=True)

    # --- Summary print ---
    print(f"Current price: ${current_market_price:,.2f}")
    print(f"Open shares: {open_quantity:.0f}")
    print(f"Avg cost (open): ${average_cost_basis:,.2f}")
    print(f"Market value: ${market_value:,.2f}")
    print(f"Realized P&L (FIFO): ${cumulative_realized_pl:,.2f}")
    print(f"Unrealized P&L @ ${current_market_price}: ${unrealized_pl:,.2f}")
    print(f"Total P&L: ${total_pl:,.2f}")

    return {
        "summary": {
            "current_price": current_market_price,
            "open_qty": open_quantity,
            "avg_cost": average_cost_basis,
            "market_value": market_value,
            "realized_pl": cumulative_realized_pl,
            "unrealized_pl": unrealized_pl,
            "total_pl": total_pl,
        },
        "match_ledger": match_df,   # each sell matched to buy lots
        "trade_summary": trade_df   # realized P&L per trade
    }

# Example usage:
res = fifo_pnl(circle, 150.0)
```
